```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
pd.read_csv(r"C:\Users\Noel\Downloads\Pandas\onlinefoods1.csv")
```

```python
df= pd.read_csv(r"C:\Users\Noel\Downloads\Pandas\onlinefoods1.csv")
df
```

```python
# Define the demographic columns
demographic_columns = ['Age', 'Gender', 'Marital Status', 'Occupation', 'Educational Qualifications']

# Create subplots
fig, axes = plt.subplots(len(demographic_columns), 1, figsize=(12, 20), sharex=False)

# Iterate over each demographic column and create a bar plot
for i, column in enumerate(demographic_columns):
    # Calculate the proportion of feedback for each category
    feedback_proportions = pd.crosstab(df[column], df['Feedback'], normalize='index')
    
    # Plot the bar plot
    feedback_proportions.plot(kind='bar', stacked=True, ax=axes[i], color=['#1f77b4', '#ff7f0e'], alpha=0.7)
    
    # Set labels and title
    axes[i].set_title(f'Feedback by {column}')
    axes[i].set_ylabel('Proportion')
    axes[i].set_xlabel(column)
    axes[i].legend(title='Feedback')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Filter the DataFrame to include only rows where Feedback is 'Positive'
positive_feedback_df = df[df['Feedback'] == 'Positive']

# Create a scatter plot for Positive feedback only
plt.figure(figsize=(12, 6))
sns.scatterplot(data=positive_feedback_df, x='longitude', y='latitude', color='red')

# Set labels and title
plt.title('Customer Locations with Positive Feedback')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show plot
plt.show()
```

```python
# Define the demographic columns
demographic_columns = ['Age', 'Gender', 'Marital Status', 'Occupation', 'Educational Qualifications']

# Set up the plot
fig, axes = plt.subplots(len(demographic_columns), 1, figsize=(12, 25), sharex=False)

# Iterate over each demographic column and create a bar plot
for i, column in enumerate(demographic_columns):
    # Calculate the count of 'Yes' and 'No' in Output for each category
    order_counts = pd.crosstab(df[column], df['Output'], normalize='index')
    
    # Plot the bar plot
    order_counts.plot(kind='bar', stacked=True, ax=axes[i], color=['#1f77b4', '#ff7f0e'], alpha=0.7)
    
    # Set labels and title
    axes[i].set_title(f'Order Frequency by {column}')
    axes[i].set_ylabel('Proportion')
    axes[i].legend(title='Output', loc='upper right')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
